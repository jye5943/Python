## 2. xml데이터 불러오기

```python
request = urllib.request.Request(xmlurl) 
request.get_method = lambda: 'GET' 
response_body = urllib.request.urlopen(request).read() 
fullXmlString = str(response_body, "utf-8") #utf-8 로 encoding 하여 str 형태로 받겠다
```

## 3. 데이터 조회에 필요한 lib import

```python
import xml.etree.ElementTree as ET
from xml.etree.ElementTree import parse, Element
```

```python
result = []
```

```python
# xml 라이브러리 표준
root = ET.fromstring(fullXmlString) #extended elements로 재구성을 해서 root에 저장한다.
# body->items->item 모두 선택
elements = root.findall('body/items/item')
```

```python
elements[0].find('거래금액').text
```

- 디버깅 코드

```python
result2 = []
item = elements[0]
item_list = []
item_list.append(item.find('거래금액').text)
item_list.append(item.find('건축년도').text)
result2.append(item_list)
```

```python
result2
```

- 디버깅 성공해서 for문 코딩

```python
for item in elements:
    try:
        item_list = []
        item_list.append(item.find('거래금액').text)
        item_list.append(item.find('건축년도').text)
        result.append(item_list)
    except Exception as e:
        print("This row will be ignored. ",item_list) 
```

```python
result
```

- pandas를 이용해 datafram화 하기

```python
finalResult = pd.DataFrame(result)
finalResult.columns = ["거래금액","건축년도"]
finalResult.head()
```
